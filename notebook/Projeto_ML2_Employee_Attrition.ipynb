```python
# Tratando valores nulos
df_clean = df.dropna()

# Codificando variáveis categóricas usando o método get_dummies
df_encoded = pd.get_dummies(df_clean, columns=['dept', 'salary'])

# Verificando o resultado do pré-processamento
df_encoded.head()
df_encoded.info()
```

>## Categorização da coluna "satisfaction_level".



```python
# Categorizando a coluna 'satisfaction_level'
# Definindo os limites para baixa, média e alta satisfação
bins = [0, 0.33, 0.66, 1]
labels = ['Baixa', 'Média', 'Alta']
df_encoded['satisfaction_category'] = pd.cut(df_encoded['satisfaction_level'], bins=bins, labels=labels, include_lowest=True)

# Separando características (features) e rótulos (labels)
X = df_encoded.drop(['Emp ID', 'satisfaction_level', 'satisfaction_category'], axis=1)
y = df_encoded['satisfaction_category']

# Visualizando a distribuição das categorias
category_distribution = y.value_counts()
category_distribution
```

A coluna "satisfaction_level" foi categorizada com sucesso em três classes: "Baixa", "Média" e "Alta". A distribuição das categorias é a seguinte:

* Alta: 7.051
* Média: 5.862
* Baixa: 2.086

Os dados foram divididos com sucesso em conjuntos de treino e teste. O conjunto de treino possui 10.499 amostras e o conjunto de teste 4.500 amostras, cada um com 19 características (features).

## Código para SVM com Pipeline e GridSearchCV

```python
%%time
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Função para avaliar o modelo
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)

    print(f"Acurácia: {accuracy}\n")
    print("Relatório de Classificação:\n", report)

    # Exibindo a matriz de confusão
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predito')
    plt.ylabel('Verdadeiro')
    plt.title('Matriz de Confusão')
    plt.show()

# Criando o pipeline para o modelo SVM
pipeline_svm = Pipeline([
    ('scaler', StandardScaler()),  # Padronização dos dados
    ('svm', SVC())  # Modelo SVM
])

# Definindo o espaço de parâmetros para o GridSearchCV
param_grid_svm = {
    'svm__C': [0.1, 1, 10],  # Valores de C para regularização
    'svm__gamma': [1, 0.1, 0.01],  # Valores de gamma
    'svm__kernel': ['rbf', 'linear']  # Tipos de kernel
}

# Criando o GridSearchCV
grid_search_svm = GridSearchCV(pipeline_svm, param_grid_svm, cv=5, verbose=2, n_jobs=-1)

# Treinando o modelo usando o conjunto de treino
grid_search_svm.fit(X_train, y_train)

# Avaliando o modelo no conjunto de teste
evaluate_model(grid_search_svm, X_test, y_test)

# Melhores parâmetros e pontuação
print("Melhores parâmetros:", grid_search_svm.best_params_)
print("Melhor pontuação no treinamento:", grid_search_svm.best_score_)
```

Este código cria um pipeline para o modelo SVM, padronizando os dados antes de aplicar o modelo. O GridSearchCV é utilizado para buscar os melhores parâmetros para o modelo dentro do espaço de parâmetros definido. É importante ajustar os valores dos parâmetros e o número de folds de validação cruzada (cv) conforme necessário

## Código para RandomForestClassifier com Avaliação do Modelo

```python
from sklearn.ensemble import RandomForestClassifier

# Criando o pipeline para o modelo RandomForestClassifier
pipeline_rf = Pipeline([
    ('scaler', StandardScaler()),  # Padronização dos dados
    ('rf', RandomForestClassifier())  # Modelo RandomForest
])

# Definindo o espaço de parâmetros para o GridSearchCV
param_grid_rf = {
    'rf__n_estimators': [50, 100, 200],  # Número de árvores
    'rf__max_depth': [None, 10, 20, 30],  # Profundidade máxima da árvore
    'rf__min_samples_split': [2, 5, 10]  # Número mínimo de amostras para dividir um nó
}

# Criando o GridSearchCV
grid_search_rf = GridSearchCV(pipeline_rf, param_grid_rf, cv=5, verbose=2, n_jobs=-1)

# Treinando o modelo usando o conjunto de treino
grid_search_rf.fit(X_train, y_train)

# Avaliando o modelo no conjunto de teste
evaluate_model(grid_search_rf, X_test, y_test)

# Melhores parâmetros e pontuação
print("Melhores parâmetros:", grid_search_rf.best_params_)
print("Melhor pontuação no treinamento:", grid_search_rf.best_score_)
```

## Código para AdaBoost com Avaliação do Modelo

```python
from sklearn.ensemble import AdaBoostClassifier

# Criando o pipeline para o modelo AdaBoost
pipeline_adaboost = Pipeline([
    ('scaler', StandardScaler()),  # Padronização dos dados
    ('adaboost', AdaBoostClassifier())  # Modelo AdaBoost
])

# Definindo o espaço de parâmetros para o GridSearchCV
param_grid_adaboost = {
    'adaboost__n_estimators': [50, 100, 200],  # Número de árvores
    'adaboost__learning_rate': [0.01, 0.1, 1]  # Taxa de aprendizado
}

# Criando o GridSearchCV
grid_search_adaboost = GridSearchCV(pipeline_adaboost, param_grid_adaboost, cv=5, verbose=2, n_jobs=-1)

# Treinando o modelo usando o conjunto de treino
grid_search_adaboost.fit(X_train, y_train)

# Avaliando o modelo no conjunto de teste
evaluate_model(grid_search_adaboost, X_test, y_test)

# Melhores parâmetros e pontuação
print("Melhores parâmetros:", grid_search_adaboost.best_params_)
print("Melhor pontuação no treinamento:", grid_search_adaboost.best_score_)
```

## Código para XGBoost com Avaliação do Modelo

>### Código para DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors

# Supondo que X_scaled é o seu conjunto de dados padronizado
# e min_samples é o mesmo valor que você usaria para o DBSCAN

# Escolha de min_samples normalmente é o mesmo que o DBSCAN ou 1 mais que ele se contar o próprio ponto
min_samples = 100

# Criando o objeto NearestNeighbors
nearest_neighbors = NearestNeighbors(n_neighbors=min_samples)
nearest_neighbors.fit(X_scaled)
distances, indices = nearest_neighbors.kneighbors(X_scaled)

# Ordenando as distâncias
sorted_distances = np.sort(distances[:, min_samples-1], axis=0)

# Plotando o gráfico de distância k-NN
plt.figure(figsize=(10, 6))
plt.plot(sorted_distances)
plt.title('Gráfico de distância k-NN')
plt.xlabel('Ponto')
plt.ylabel('Distância para o k-ésimo vizinho mais próximo')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()



```

```python
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA


# Definir os parâmetros para o DBSCAN com base na análise inicial
eps_value = 3  # Valor inicial baseado no gráfico de distância k-NN
min_samples_value = 3 * X_scaled.shape[1]  # O dobro do número de características

# Criar o modelo DBSCAN com os parâmetros definidos
dbscan = DBSCAN(eps=eps_value, min_samples=min_samples_value)
dbscan.fit(X_scaled)

# Obter os rótulos dos clusters
clusters_dbscan = dbscan.labels_

# Realizar PCA para a visualização em 2D
pca = PCA(n_components=7)
X_pca = pca.fit_transform(X_scaled)

# Visualizar os clusters
plt.figure(figsize=(10, 10))
unique_labels = set(clusters_dbscan)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Preto usado para ruído.
        col = [0, 0, 0, 1]

    class_member_mask = (clusters_dbscan == k)

    xy = X_pca[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14 if k == -1 else 6)

plt.title('Clusters DBSCAN com PCA')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.show()

# Retornar a quantidade de clusters encontrados e os pontos considerados ruído
(unique, counts) = np.unique(clusters_dbscan, return_counts=True)
clusters_counts = dict(zip(unique, counts))

clusters_counts
```
